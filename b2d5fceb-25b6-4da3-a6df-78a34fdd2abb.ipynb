```python
# Import `load_workbook` module from `openpyxl`
from openpyxl import load_workbook
import datetime
import numpy as np
import matplotlib.pyplot as plt



# Load in the workbook
wb = load_workbook('./tink1.xlsx')

# Get sheet names
#print(wb.sheetnames)
sheet = wb['Лист1']
#sheet['A1'].value
reg_dttm =[[(cell.value).strftime("%m") for cell in cellObj] for cellObj in sheet['C2':'C2378'] ]
birth_dt =[[(cell.value).strftime("%Y") for cell in cellObj] for cellObj in sheet['B2':'B2378'] ] 
#int
reg_dttm_int = [int(item[0]) for item in reg_dttm]
birth_dt_int = [2020-int(item[0]) for item in birth_dt]
#list whis summ of consumers per month
n1=0
n2=0
n3=0
n4=0
for i in reg_dttm_int:
    if i==1:
        n1=n1+1# 710
    if i==2:
        n2=n2+1#942
    if i==3:
        n3=n3+1#674
    if i==4:
        n4=n4+1#51
#array with age consumers in current month
reg_1=[birth_dt_int[i] for i in range(0,710)]
reg_2=[birth_dt_int[i] for i in range(710,711+941)]
reg_3=[birth_dt_int[i] for i in range(711+941,711+941+674)]
reg_4=[birth_dt_int[i] for i in range(711+941+674,711+941+674+51)]

a1 = []
a2 = []
a3 = []
a4 = []
for i in range(0,100):
    a1.append(0)
    a2.append(0)
    a3.append(0)
    a4.append(0)
for i in reg_1:
    a1[i]=a1[i]+1
for i in reg_2:
    a2[i]=a2[i]+1
for i in reg_3:
    a3[i]=a3[i]+1
for i in reg_4:
    a4[i]=a4[i]+1
a11=[a1[i] for i in range(18,75)]
a22=[a2[i] for i in range(18,75)]
a33=[a3[i] for i in range(18,75)]
a44=[a4[i] for i in range(18,75)] 
stolb_diag4(a22, a11,a33,a44)  
```
