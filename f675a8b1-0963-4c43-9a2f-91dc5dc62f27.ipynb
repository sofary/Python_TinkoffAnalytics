```python
def diag(x,y):
    fig, ax = plt.subplots()
    ax.plot(x, y)
    ax.set_facecolor('seashell')
    fig.set_facecolor('floralwhite')
    ax.set_ylabel('average summ of first purchase, days')
    ax.set_xlabel('utilizations speed, dollars')
    fig.set_figwidth(12)    #  ширина Figure
    fig.set_figheight(6)    #  высота Figure
    plt.show()
```

```python
from openpyxl import load_workbook
import datetime
import numpy as np
import matplotlib.pyplot as plt
import pandas 
import math


excel_d = pandas.read_excel('./tink1.xlsx', sheet_name='Лист1', usecols=['purchase_dt', 'reg_dttm','purchase_amt'])
a=[i for i in range(0, 2376)  if math.isnan(excel_d.iat[i,2])]
excel_d=excel_d.drop(a,axis=0)
speed=[(excel_d.iat[i,1] -excel_d.iat[i,0]).days  for i in range(0, excel_d.shape[0])]
cash=[excel_d.iat[i,2] for i in range(0, excel_d.shape[0])]
us=[[speed[i],int(cash[i])] for i in range(0, excel_d.shape[0]-1)]
sorted(us, key=lambda x: x[0], reverse=True)
s=[[0,0,i] for i in range(0,88)]#0 - число раз, 1 - общая сумма
for item in us:
    s[item[0]][0]=s[item[0]][0]+1
    s[item[0]][1]=s[item[0]][0] +item[1]
ss=[[item[1]/item[0], item[2]] for item in s if item[0]!=0]#массив средней цены по повторяющимся скоростям
s1=[item[1] for item in ss]
s0=[item[0] for item in ss]
diag(s1,s0)
```
